## Part 1: USSD Banking App

This mini USSD app demonstrates the use of:

- Variables
- Conditional statements
- Loops
- Functions
- Basic input handling

The selected bank for this demo is **IQ Bank**.

```python
# IQ Bank USSD App
# This program demonstrates variables, loops, functions, and conditional statements.

bank_name = "IQ Bank"
account_name = "Isaac Ogbaji"
account_number = "0123456789"
balance = 50000.00
pin = "1234"
airtime_networks = ["MTN", "Airtel", "Glo", "9mobile"]
transaction_history = []


def verify_pin():
    """Ask the user to enter PIN and verify it."""
    entered_pin = input("Enter your 4-digit PIN: ")
    return entered_pin == pin


def show_balance():
    """Display account balance."""
    print(f"\n{bank_name} Balance Enquiry")
    print(f"Account Name: {account_name}")
    print(f"Account Number: {account_number}")
    print(f"Available Balance: ₦{balance:,.2f}")


def transfer_money():
    """Transfer money to another account."""
    global balance
    print("\nTransfer Money")
    recipient_account = input("Enter recipient account number: ")
    recipient_bank = input("Enter recipient bank name: ")
    amount = float(input("Enter amount to transfer: ₦"))

    if amount <= 0:
        print("Invalid amount. Please enter a positive value.")
    elif amount > balance:
        print("Insufficient balance.")
    elif verify_pin():
        balance -= amount
        transaction_history.append(f"Transferred ₦{amount:,.2f} to {recipient_account} - {recipient_bank}")
        print(f"Transfer successful! New balance: ₦{balance:,.2f}")
    else:
        print("Incorrect PIN. Transaction failed.")


def buy_airtime():
    """Buy airtime for self or another phone number."""
    global balance
    print("\nBuy Airtime")
    print("Select Network:")

    for index, network in enumerate(airtime_networks, start=1):
        print(f"{index}. {network}")

    network_choice = int(input("Enter network option: "))

    if network_choice < 1 or network_choice > len(airtime_networks):
        print("Invalid network option.")
        return

    selected_network = airtime_networks[network_choice - 1]
    phone_number = input("Enter phone number: ")
    amount = float(input("Enter airtime amount: ₦"))

    if amount <= 0:
        print("Invalid amount.")
    elif amount > balance:
        print("Insufficient balance.")
    elif verify_pin():
        balance -= amount
        transaction_history.append(f"Bought ₦{amount:,.2f} airtime for {phone_number} on {selected_network}")
        print(f"Airtime purchase successful! New balance: ₦{balance:,.2f}")
    else:
        print("Incorrect PIN. Transaction failed.")


def deposit_money():
    """Deposit money into the account."""
    global balance
    print("\nDeposit Money")
    amount = float(input("Enter deposit amount: ₦"))

    if amount <= 0:
        print("Invalid deposit amount.")
    else:
        balance += amount
        transaction_history.append(f"Deposited ₦{amount:,.2f}")
        print(f"Deposit successful! New balance: ₦{balance:,.2f}")


def show_transaction_history():
    """Display all transactions made in the current session."""
    print("\nTransaction History")

    if len(transaction_history) == 0:
        print("No transactions yet.")
    else:
        for number, transaction in enumerate(transaction_history, start=1):
            print(f"{number}. {transaction}")


def ussd_menu():
    """Main USSD menu loop."""
    while True:
        print(f"\nWelcome to {bank_name} USSD Banking")
        print("1. Check Balance")
        print("2. Transfer Money")
        print("3. Buy Airtime")
        print("4. Deposit Money")
        print("5. Transaction History")
        print("6. Exit")

        choice = input("Select an option: ")

        if choice == "1":
            if verify_pin():
                show_balance()
            else:
                print("Incorrect PIN.")
        elif choice == "2":
            transfer_money()
        elif choice == "3":
            buy_airtime()
        elif choice == "4":
            deposit_money()
        elif choice == "5":
            show_transaction_history()
        elif choice == "6":
            print(f"Thank you for using {bank_name} USSD Banking. Goodbye!")
            break
        else:
            print("Invalid option. Please try again.")


# Run the USSD app
# Default PIN for testing: 1234
ussd_menu()
```

## Part 3: Load Sample Data with Pandas and NumPy

The sample data below represents bank customer transactions.

```python
import pandas as pd
import numpy as np

# Create sample customer transaction data
sample_data = {
    "Customer_ID": [101, 102, 103, 104, 105],
    "Customer_Name": ["Isaac", "Ada", "Chinedu", "Amina", "Tunde"],
    "Transaction_Type": ["Deposit", "Transfer", "Airtime", "Withdrawal", "Deposit"],
    "Amount": [50000, 15000, 2000, 10000, 30000],
    "Account_Balance": [120000, 85000, 43000, 60000, 95000]
}

# Load the data into a Pandas DataFrame
df = pd.DataFrame(sample_data)

# Display the DataFrame
df
```

```python
# Basic data analysis using Pandas and NumPy

print("First 5 Records:")
print(df.head())

print("\nData Summary:")
print(df.describe())

print("\nTotal Transaction Amount:")
print(np.sum(df["Amount"]))

print("\nAverage Transaction Amount:")
print(np.mean(df["Amount"]))

print("\nHighest Transaction Amount:")
print(np.max(df["Amount"]))

print("\nLowest Transaction Amount:")
print(np.min(df["Amount"]))
```
